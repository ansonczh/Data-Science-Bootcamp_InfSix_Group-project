```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame manually based on the image
data = {
    'Platform': [
        'Twitter', 'Twitter', 'Instagram', 'Facebook', 'Instagram',
        'Twitter', 'Facebook', 'Instagram', 'Twitter', 'Facebook'
    ],
    'Sentiment': [
        'Positive', 'Negative', 'Positive', 'Positive', 'Neutral',
        'Positive', 'Positive', 'Positive', 'Negative', 'Neutral'
    ]
}

df = pd.DataFrame(data)

# Count posts per platform
platform_counts = df['Platform'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
platform_counts.plot(kind='bar', color='lightcoral')

plt.title('Number of Posts per Platform (Top 10 Data)')
plt.xlabel('Platform')
plt.ylabel('Number of Posts')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Recreate data based on your screenshot
data = {
    'Platform': [
        'Twitter', 'Twitter', 'Instagram', 'Facebook', 'Instagram',
        'Twitter', 'Facebook', 'Instagram', 'Twitter', 'Facebook'
    ],
    'Sentiment': [
        'Positive', 'Negative', 'Positive', 'Positive', 'Neutral',
        'Positive', 'Positive', 'Positive', 'Negative', 'Neutral'
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Group by Platform and Sentiment and count occurrences
sentiment_counts = df.groupby(['Platform', 'Sentiment']).size().unstack(fill_value=0)

# Plot grouped bar chart
sentiment_counts.plot(kind='bar', figsize=(10, 6))

plt.title('Sentiment Distribution per Platform (Top 10 Data)')
plt.xlabel('Platform')
plt.ylabel('Number of Posts')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data based on your screenshot
data = {
    'Country': [
        'USA', 'Canada', 'USA', 'UK', 'Australia',
        'India', 'Canada', 'USA', 'USA', 'Australia'
    ],
    'Sentiment': [
        'Positive', 'Negative', 'Positive', 'Positive', 'Neutral',
        'Positive', 'Positive', 'Positive', 'Negative', 'Neutral'
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Get list of unique countries
countries = df['Country'].unique()

# Plot a pie chart for each country
fig, axes = plt.subplots(nrows=2, ncols=(len(countries) + 1) // 2, figsize=(14, 8))

for i, country in enumerate(countries):
    ax = axes.flat[i]
    country_data = df[df['Country'] == country]['Sentiment'].value_counts()
    
    ax.pie(
        country_data,
        labels=country_data.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Set3.colors
    )
    ax.set_title(f'Sentiment in {country}')

# Hide any unused subplots
for j in range(i + 1, len(axes.flat)):
    axes.flat[j].axis('off')

plt.suptitle('Sentiment Distribution by Country (Top 10 Posts)', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()

```
