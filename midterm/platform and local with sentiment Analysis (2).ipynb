```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("data.csv", encoding="ISO-8859-1")

# Define the sentiments to include
target_sentiments = ['Positive', 'Negative', 'Happy', 'Joy', 'Sad', 'Hopeful']

# Filter the dataset to only include rows with these sentiments
filtered_df = df[df['Sentiment'].isin(target_sentiments)]

# Group by Country and Sentiment and count the number of occurrences
sentiment_by_country = filtered_df.groupby(['Country', 'Sentiment']).size().unstack(fill_value=0)

# Plotting the bar chart
sentiment_by_country.plot(kind='bar', figsize=(12, 6), width=0.75)

plt.title('Sentiment Distribution by Country (Positive, Negative, Joy, etc.)')
plt.xlabel('Country')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Sentiment')
plt.tight_layout()

plt.show()
```
